````
def entropy(matches):
    count_dic={"A":0,"G":0,"C":0,"T":0, "a":0, "t":0, "g":0, "c": 0}
    entropy=0
    for i in matches:
      if i in count_dic.keys():
          count_dic[i] += 1
    base_count = sum(count_dic.values())
    for base, count in count_dic.items():
        if count !=0:
            entropy += (count/base_count)*np.log(count/base_count)
    return -entropy

def base_count(matches):
    count_dic={"A":0,"G":0,"C":0,"T":0, "a":0, "t":0, "g":0, "c": 0}
    entropy=0
    for i in matches:
      if i in count_dic.keys():
          count_dic[i] += 1
    return sum(count_dic.values())
````

````
toremove = re.compile('[<>$*#^]')
pileup_new=None
for df in pileup:
    df['matches'] = df['basereads'].apply(lambda x: toremove.sub('', x))
    df['base count'] = df.apply(lambda x : base_count(str(x["matches"])), axis=1)
    df["entropy"] = df.apply(lambda x : entropy(str(x["matches"])), axis=1)
    pileup_new = pd.concat([pileup_new, df], axis=0)
````

# Data preprocessing

````
def read2matrix(read):
    matrix = np.zeros((4,18))
    for i in range(len(read)):
        if read[i]=='A': matrix[0][i]=1
        elif read[i]=='G': matrix[1][i]=1
        elif read[i]=='C': matrix[2][i]=1
        elif read[i]=='T': matrix[3][i]=1
        elif read[i]=='N': matrix[:,i]=0.25
    return matrix
````

````
!wget GRCm39.genome.fa.gz https://ftp.ebi.ac.uk/pub/databases/gencode/Gencode_mouse/release_M29/GRCm39.genome.fa.gz
!gunzip GRCm39.genome.fa.gz
````

````
#pileup_new_pos=pd.read_csv('pileup_new_pos.csv')
pileup_bed = pileup_new_pos[["chrom", "anterior pos", "posterior pos"]]
pileup_bed.to_csv("pileup_bed.BED", sep='\t', index=False, header=False)
!bedtools getfasta -tab -fo ./"read_data.csv" -fi ./GRCm39.genome.fa -bed "pileup_bed.BED"
````

````
Ref_data = pd.read_csv("read_data.csv", sep="\t", names=["location", "sequence"])
Ref_data["matrix"]=Ref_data["sequence"].apply(lambda x : read2matrix(x))
Ref_data["decoy sequence"] = Ref_data["sequence"].apply(lambda x : "".join(random.sample(x,len(x))))
Ref_data["decoy matrix"] = Ref_data["decoy sequence"].apply(lambda x : read2matrix(x))
Ref_data
````

````
pos_ref_dataset = np.zeros((len(Ref_data),4,18))
for i in range(len(Ref_data)):
    pos_ref_dataset[i] =Ref_data["matrix"][i]
print(pos_ref_dataset.shape)

neg_ref_dataset = np.zeros((len(Ref_data),4,18))
for i in range(len(Ref_data)):
    neg_ref_dataset[i] =Ref_data["decoy matrix"][i]
print(neg_ref_dataset.shape)
````

````
pos_ref_label = np.zeros((len(pos_ref_dataset),2))
pos_ref_label[:,1]=1
neg_ref_label = np.zeros((len(neg_ref_dataset),2))
neg_ref_label[:,0]=1
ref_data, ref_label = np.concatenate((pos_ref_dataset,neg_ref_dataset),axis=0), np.concatenate((pos_ref_label,neg_ref_label),axis=0)
ref_mask = (list(range(len(pos_ref_dataset)*2)))
random.shuffle(ref_mask)
ref_data_shuffled, ref_label_shuffled = ref_data[ref_mask], ref_label[ref_mask]
np.save("ref_data_shuffled.npy", ref_data_shuffled)
np.save("ref_label_shuffled.npy", ref_label_shuffled)
````

````
ref_data_shuffled=np.load("ref_data_shuffled.npy")
ref_label_shuffled=np.load("ref_label_shuffled.npy")
````

````
ref_dataset_size = ref_data_shuffled.shape[0]
TRAIN_RATE = 0.8
Train_size = int(ref_dataset_size*TRAIN_RATE)
Valid_size = int((ref_dataset_size-Train_size)/2)
Test_size = ref_dataset_size-(Train_size+Valid_size)
Train_ref_set, Train_ref_label = ref_data_shuffled[:Train_size], ref_label_shuffled[:Train_size]
Valid_ref_set, Valid_ref_label = ref_data_shuffled[Train_size:Train_size+Valid_size], ref_label_shuffled[Train_size:Train_size+Valid_size]
Test_ref_set, Test_ref_label = ref_data_shuffled[Train_size+Valid_size:], ref_label_shuffled[Train_size+Valid_size:]
````

# Deep learning training

````
tf.random.set_seed(42)
model = None
````

````
# CNN-based
model=tf.keras.models.Sequential()
model.add(tf.keras.layers.Conv2D(16,7,activation="relu",padding="same", input_shape = [4,18,1], kernel_initializer="he_normal"))
model.add(tf.keras.layers.MaxPooling2D(2))
model.add(tf.keras.layers.Conv2D(16,7,activation="relu",padding="same", kernel_initializer="he_normal"))
model.add(tf.keras.layers.MaxPooling2D(2))
model.add(tf.keras.layers.Flatten())
model.add(tf.keras.layers.Dense(64, kernel_initializer="he_normal"))
model.add(tf.keras.layers.LeakyReLU(alpha=0.2))
model.add(tf.keras.layers.Dropout(0.2))
model.add(tf.keras.layers.Dense(32, kernel_initializer="he_normal"))
model.add(tf.keras.layers.LeakyReLU(alpha=0.2))
model.add(tf.keras.layers.Dropout(0.2))
model.add(tf.keras.layers.Dense(2,activation="sigmoid"))

optimizer = tf.keras.optimizers.SGD(learning_rate=0.1)
model.compile(loss="binary_crossentropy",optimizer=optimizer, metrics=["accuracy"])

checkpoint_cb = tf.keras.callbacks.ModelCheckpoint("Best_model_ref_CNN.h5", save_best_only=True)
early_stopping_cb = tf.keras.callbacks.EarlyStopping(patience = 10, restore_best_weights=True)

history = model.fit(Train_ref_set, Train_ref_label, epochs=40, validation_data=(Valid_ref_set,Valid_ref_label), callbacks=[checkpoint_cb, early_stopping_cb])

#model.load_weights("Best_model_ref.h5")
````

````
model.evaluate(Test_ref_set, Test_ref_label)
print(roc_auc_score(Test_ref_label, model.predict(Test_ref_set)))
````

````
def matrix2read(matrix):
    read_list=[]
    for i in range(matrix.shape[0]):
        read_matrix = matrix[i]
        read=""
        for j in range(read_matrix.shape[1]):
            if int(read_matrix[0][j])==1 : read += "A"
            elif int(read_matrix[1][j])==1 : read += "G"
            elif int(read_matrix[2][j])==1 : read += "C"
            elif int(read_matrix[3][j])==1 : read += "T"
            elif read_matrix[0][j] == 0.25 : read += "N"
        read_list.append(read)
    return read_list
````

````
def measure_acc(testset, label_num, active_model, inactive_model):
    total = 0
    for i in range(len(testset)):
        score_arg = np.argmax([inactive_model.score(testset[i].reshape(1,-1)), active_model.score(testset[i].reshape(1,-1))])
        if score_arg==label_num: total+=1
    return total/len(testset)*100
````
